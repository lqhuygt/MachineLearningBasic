```python
#import thư viện
import sklearn.datasets as datask
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
#đọc dữ liệu
X_diabetes, y_diabetes = datask.load_diabetes(return_X_y = True)
print(type(X_diabetes))
print("kích thước tập train: ", X_diabetes.shape)
print("kích thước tập test: ", y_diabetes.shape)
```

```python
#phân tích dữ liệu
x_features = pd.DataFrame(X_diabetes, columns = ['age','sex','bmi','bp','s1 tc','s2 ldl','s3 hdl','s4 tch','s5 ltg','s6 glu'])
x_features.head()
```

```python
x_features.describe()
```

```python
# quan sát đặc trưng của target
y_label = pd.DataFrame(y_diabetes, columns=['qmodp'])
y_label.describe()
```

```python
#ghép input với output vào 1 dataframe
diabetes = pd.concat([x_features,y_label],axis=1)
diabetes.head()
```

```python
#tách tập train và test 75-25
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_features,y_label, test_size=0.2,random_state=50)
print("kích thước tập train: ", len(y_train))
print("kích thước tập test: ", len(y_test))
```

```python
#train mô hình
from sklearn.linear_model import LinearRegression

lin_model = LinearRegression(fit_intercept=True)
lin_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# dự đoán
y_pred = lin_model.predict(X_test)
# đánh giá độ chính xác của mô hình
rmse = (np.sqrt(mean_squared_error(y_pred,y_test)))
print(rmse)
r2 = r2_score(y_pred,y_test)
print(r2)
```
