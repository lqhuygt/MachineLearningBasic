```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime 
```

```python
df = pd.read_csv("ted_main.csv")
df.head()
```

```python
# convert date về định dang d-m-y
df['film_date'] = df['film_date'].apply(lambda x: datetime.datetime.fromtimestamp(int(x)).strftime('%d-%m-%Y'))
df['published_date'] = df['published_date'].apply(lambda x: datetime.datetime.fromtimestamp(int(x)).strftime('%d-%m-%Y'))
df.head(1)
```

```python
# convert thời lượng video về số phút
df['duration'] = df['duration'] / 60
df.head(1)
```

### Top 15 bài nói phổ biến mọi thời đại

```python
# dựa vào số lượt xem
pop_views = df[['title', 'main_speaker', 'views', 'film_date']].sort_values('views', ascending=False)[:15]
# pop_talks.head(15)
pop_views
```

```python
pop_views['main_spk'] = pop_views['main_speaker'].apply(lambda x: x[:3]) # lấy 3 kí tự đầu tiên từ tên người nói
pop_views.plot.bar(rot=0, figsize=(10,6), x='main_spk', y='views' );
plt.ylabel("trục triệu");
```

```python
# cách khác dùng sea born đẻ plot
import seaborn as sns

pop_views['main_spk'] = pop_views['main_speaker'].apply(lambda x: x[:3])
sns.set_style("whitegrid")
plt.figure(figsize=(15,6))
sns.barplot(x='main_spk', y='views', data=pop_views);
```

```python
# dựa vào lượt comment
pop_comments = df[['title', 'main_speaker', 'views', 'comments']].sort_values('comments', ascending=False)[:15]
# pop_talks.head(15)
pop_comments
```

### Tháng nào có các bài nói được đăng nhiều nhất

```python
# tạo cột tháng phânt tách từ cột fim_date
df['month'] = df['film_date'].apply(lambda x: int(x.split('-')[1]))
df_months = df['month'].value_counts().sort_index()
df_months
```

```python
df_months.plot.bar(rot=0, figsize=(10,6))
plt.xlabel("Tháng")
plt.ylabel("Takls");
```

```python
# dùng sns để plot
month_df = pd.DataFrame(df['month'].value_counts()).reset_index()
month_df.columns = ['month', 'talks']
plt.figure(figsize=(10,6))
sns.barplot(x='month', y='talks', data=month_df)
```

### Thứ mấy trong tuần có bài đăng nhiều nhất

```python
# hàm lấy thứ trong tuần
def getday(x):
    day, month, year = (int(i) for i in x.split('-'))
    answer = datetime.date(year, month, day)
    return (answer.strftime('%A'))
```

```python
# tạo cột thứ
df['day'] = df['film_date'].apply(getday)
df.head()
```

```python
df['day'].dtype
```

```python
pop_day = df['day'].value_counts()
pop_day
```

```python
pop_day.plot.bar(rot=20,figsize = (10,6));
plt.xlabel("Thứ", fontsize=13)
```
