### **Yêu cầu 2.2:**
---
Nối 3 DataFrame df_lop1, df_lop2, df_lop3 thành một DataFrame df_full chứa tất cả danh sách bảng điểm của 3 lớp

### **Yêu cầu 2.3:**
---
Trong df_full: Tạo một cột Diem_he10 được tính dựa vào các cột tương ứng của từng hàng dữ liệu, theo công thức sau:

**Diem_he10 = 0.6 * A + 0.3 * ((B1+B2)/2) + 0.1 * ((C1+C2)/2)**
    
Làm tròn đến 1 số sau dấu phẩy

```python
#hàm tính điểm hệ 10
def diemhe10(a,b1,b2,c1,c2):
    diem = round(0.6 * a + 0.3 * ((b1+b2)/2) + 0.1 * ((c1+c2)/2),1)
    return diem
```

```python
df_full['Diem_he10'] = df_full.apply(lambda row: diemhe10(row['A'],row['B1'],row['B2'],row['C1'],row['C2']), axis=1)
df_full
```

### **Yêu cầu 2.4:**
---
Từ cột Diem_he10 trong df_full tạo một cột Diem_chu, Diem_so theo quy đổi:

    * Điểm hệ 10 Từ 9.0 đến 10.0  ----> Điểm chữ: A+  ----> Điểm số: 4.0
    * Điểm hệ 10 Từ 8.5 đến 8.9  ----> Điểm chữ: A    ----> Điểm số: 3.7
    * Điểm hệ 10 Từ 8.0 đến 8.4  ----> Điểm chữ: B+   ----> Điểm số: 3.5   
    * Điểm hệ 10 Từ 7.0 đến 7.9  ----> Điểm chữ: B    ----> Điểm số: 3.0
    * Điểm hệ 10 Từ 6.5 đến 6.9  ----> Điểm chữ: C+   ----> Điểm số: 2.5  
    * Điểm hệ 10 Từ 5.5 đến 6.4  ----> Điểm chữ: C    ----> Điểm số: 2.0     
    * Điểm hệ 10 Từ 5.0 đến 5.4  ----> Điểm chữ: D+   ----> Điểm số: 1.5 
    * Điểm hệ 10 Từ 4.0 đến 4.9  ----> Điểm chữ: D    ----> Điểm số: 1.0  
    * Điểm hệ 10 Từ 0.0 đến 3.9  ----> Điểm chữ: F    ----> Điểm số: 0.0 

```python
def check_diemchu(x):
    if (9.0 <= x <= 10.0 ):
        return 'A+'
    elif (8.5 <= x <= 8.9):
        return 'A'
    elif (8.0 <= x <= 8.4):
        return 'B+'
    elif (7.0 <= x <= 7.9):
        return 'B'
    elif (6.5 <= x <= 6.9):
        return 'C+'
    elif (5.5 <= x <= 6.4):
        return 'C'
    elif (5.0 <= x <= 5.4):
        return 'D+'
    elif (4.0 <= x <= 4.9):
        return 'D'
    else:
        return 'F'
```

```python
def check_diemso(x):
    if (9.0 <= x <= 10.0 ):
        return '4.0'
    elif (8.5 <= x <= 8.9):
        return '3.7'
    elif (8.0 <= x <= 8.4):
        return '3.5'
    elif (7.0 <= x <= 7.9):
        return '3.0'
    elif (6.5 <= x <= 6.9):
        return '2.5'
    elif (5.5 <= x <= 6.4):
        return '2.0'
    elif (5.0 <= x <= 5.4):
        return '1.5'
    elif (4.0 <= x <= 4.9):
        return '1.0'
    else:
        return '0.0'
```

```python
df_full['Diem_chu'] = df_full.apply(lambda row: check_diemchu(row['Diem_he10']), axis=1)
df_full['Diem_so'] = df_full.apply(lambda row: check_diemso(row['Diem_he10']), axis=1)
df_full
```
