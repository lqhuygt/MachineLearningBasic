### 4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (salary feature) and those who earn less than 50K per year?

```python
df['salary'].dtypes
```

```python
print("====mean====")
age_mean = df['age'].groupby(df['salary']).mean()
print(age_mean)
print("====standard deviation====")
age_std = df['age'].groupby(df['salary']).std()
print(age_std)
```

```python
d = {"mean": age_mean, "std":age_std}
mean_std_age = pd.DataFrame(data=d)
mean_std_age
```

```python
mean_std_age.plot.bar(rot=0);
```

### 9. What is the maximum number of hours a person works per week (hours-per-week feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?

```python
# số giờ tối đa một người làm việc mỗi tuần
df['hours-per-week'].max()
```

```python
# số người làm việc 99h một tuần
people_99h = df[df['hours-per-week'] == 99].value_counts().sum()
people_99h
```

```python
people_50k= df[(df['hours-per-week'] == 99) & (df['salary'] == ">50K")].value_counts().sum()
people_50k
```

```python
per = (people_50k / people_99h)*100
per
```
